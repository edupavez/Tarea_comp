```python
import numpy as np
import matplotlib as plt
from matplotlib import pyplot as pyp
import pandas as pd
```

```python
pd.read_csv("P1grupo_8.csv")
```

```python
puntos = list()

with open("P1grupo_8.csv", "r") as archivo:
    contenido = archivo.readlines()[1:]
    for linea in contenido:
        linea = linea.rstrip("\n")
        linea = linea.split(",")
        x = float(linea[0])
        y = float(linea[1])
        puntos.append((x,y))
        
```

```python
fig, ax = pyp.subplots(1, 1, figsize = (12, 12))
#ax.set_facecolor("black")
ax.grid = True

for punto in puntos:
    x = punto[0]
    y = punto[1]
    ax.scatter(x, y, marker = "o", color = "blue")
```
